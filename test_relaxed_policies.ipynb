# relaxed grid world

```julia
include("image_grid_world.jl"); # Main file includes everything else
```

## create original image grid world

```julia
igw = image_grid_world(); # This will give the default grid world we have been working with
```

## create relaxed version

```julia
rgw = RelaxedGridWorld(subset_size_reward = 1e-6);
```

## solve it using relaxed value iteration solver

```julia
solver = RelaxedValueIterationSolver(verbose = false, max_iterations = 500);
```

```julia
relaxed_policy = solve(solver, rgw);
```

## check the policy

```julia
prob = model_check(rgw, relaxed_policy, max_iterations = 100, verbose = true);
```

## visualize the results

```julia
include("support_code.jl");
```

```julia
include("viz.jl")
```

```julia
g = plot_summary(rgw, relaxed_policy, prob)
# PGFPlots.save("RelaxedPol_1em6_10000iter.pdf", g);
g
```

## fast tuning

```julia
subset_size_reward = 0.0045
γ = 0.99

rgw = RelaxedGridWorld(subset_size_reward = subset_size_reward, discount = γ)
solver = RelaxedValueIterationSolver(verbose = false, max_iterations = 500)
relaxed_policy = solve(solver, rgw)

prob = model_check(rgw, relaxed_policy, max_iterations = 100, verbose = false)

g = plot_summary(rgw, relaxed_policy, prob)
```
